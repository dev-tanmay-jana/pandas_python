```python
import pandas as pd
#to run Sift+Enter
```

```python
carsales=pd.read_csv('carsales.csv')
carsales 
```

```python
carsales['Odometer (KM)'].fillna(42000,inplace=True)
carsales
```

```python
#adding column using serise()
car_seats=pd.Series([4,4,4,5,4,6,7])
carsales['seats']=car_seats
carsales
```

```python
#fill the na value using mean()
carsales['seats']=carsales['seats'].fillna(carsales['seats'].mean())
carsales
```
