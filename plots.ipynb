```python
fig, ax=plt.subplots(1,2, figsize=(12,5))

ax[0].plot(x,y,color='g',marker='o')
ax[0].set_title('line chart')
ax[0].grid(color='r')

ax[1].bar(x,y)
ax[1].set_title('bar chart')
ax[1].grid()

plt.tight_layout()
plt.savefig('plot.png',dpi=300,bbox_inches='tight')
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix.csv')
df
```

```python
print(df.head())
```

```python
df=df.dropna(subset=['type','title','rating','ratinglevel','ratingdescription','release_year','user_rating_score','user_rating_size'])
types_count=df['type'].value_counts()
plt.Figure(figsize=(6,4))
plt.bar(types_count.index,types_count.values,color=['skyblue','orange'])
plt.title('numbber of movies vs TVshow bar chart')
plt.xlabel('types')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('TVshow_vs_movie_barchart.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
rating_count=df['rating'].value_counts()
plt.Figure(figsize=[8,6])
plt.pie(rating_count,labels=rating_count.index,autopct='%1.1f%%',startangle=90)
plt.title('ratings of content pie chart')
plt.tight_layout()
plt.savefig('content_ratings_pie.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
movie_df=df[df['type']=='movie'].copy()
movie_df['duration']=movie_df['ratingdescription'].astype(str)
plt.Figure(figsize=(8,6))
plt.hist(movie_df['duration'],bins=30,color='g',edgecolor='b')
plt.title('cistribution of movie duration')
plt.ylabel('number of movie')
plt.xlabel('duration of movie')
plt.tight_layout()
plt.savefig('movie_duration_histrogram.png')
plt.show() 
```

```python
relese_count=df['release_year'].value_counts().sort_index()
plt.Figure(figsize=(10,6))
plt.scatter(relese_count.index,relese_count.values,color='r')
plt.title('relese year vs number of shows')
plt.ylabel('number of movieshow')
plt.xlabel('relese year')
plt.tight_layout()
plt.savefig('movie_duration_scatter.png')
plt.show()
```

```python
rating_level=df['ratinglevel'].value_counts().head(10)
plt.Figure(figsize=(40,15))
plt.barh(rating_level.index,rating_level.values,color='purple')
plt.title('top 40 movies ratinglevel barh')
plt.ylabel('ratinglevel')
plt.xlabel('number of movies')
plt.tight_layout()
plt.savefig('top40_movies_barh.png')
plt.show()
```

```python
count_by_year=df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, ax=plt.subplots(1,2, figsize=(12,5))
ax[0].plot(count_by_year.index,count_by_year['movie'],color='r')
ax[0].set_title('movies per year')

ax[0].plot(count_by_year.index,count_by_year['TVshow'],color='g')
ax[0].set_title('TVshow per year')

plt.suptitle('cpmparison move vs TV show per year')
plt.tight_layout()
plt.savefig("comparison TVshow vs movie per year")
plt.show()
```
